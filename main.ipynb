```python
df['LoanAmount'].fillna(df['LoanAmount'].mean(),inplace=True)
df['LoanAmount'] = df['LoanAmount'].astype(int)
```

```python
df.head()
```

```python
#we want to convrt our object data into numeric data 
#so in 'Gender' column we perform encoding
encoder = OneHotEncoder()
encoder.fit(df[['Gender']])
Encoded_Gender = encoder.transform(df[['Gender']]).toarray()
gender_columns_names =encoder.get_feature_names_out(['Gender'])
Encoded_Gende_df = pd.DataFrame(Encoded_Gender,columns=gender_columns_names) # this is an another data frame 
Encoded_Gende_df

#this give us 3 gender but in actual we need only 2 'male' and 'female' so removeit 
Encoded_Gende_df['Gender_nan']=Encoded_Gende_df['Gender_nan'].replace(0.0,111)
Encoded_Gende_df=Encoded_Gende_df.drop(['Gender_nan'],axis=1)
Encoded_Gende_df

'''now in this we can see that there are two columns left but in our actual data frame we have only one column , 
this is the extra data which is not good .in this case we remove any one column because only one column is sufficient to 
show that a person is a male or female 
'''
Encoded_Gende_df = Encoded_Gende_df.drop(['Gender_Female'],axis=1)
Encoded_Gende_df 

#now we rename our this column from Gender_Male  ------> Gender 
Encoded_Gende_df.rename(columns={'Gender_Male':'Gender'},inplace=True)
Encoded_Gende_df  #1.0 -----------> Male and 0.0 -------->Female  
print(Encoded_Gende_df['Gender'].dtype)

#this is in flaot data type but we can convert it in int data type 
Encoded_Gende_df['Gender'] = Encoded_Gende_df['Gender'].astype(int)
print(Encoded_Gende_df['Gender'].dtype) # so now 1---> Male and 0---> female 
Encoded_Gende_df['Gender']
```

```python
#now we can extract this column into our original data frame 
df['Gender'] = Encoded_Gende_df['Gender']
```

```python
df.head()
```

```python
#now we can also apply this same process on Marridge cloumn 
encoder.fit(df[['Married']])
encoded_married = encoder.transform(df[['Married']]).toarray()
columns =encoder.get_feature_names_out(['Married'])
married_df = pd.DataFrame(encoded_married,columns=columns)
married_df

#this give us 3 married status but in actual we need only 1 to show that a person is marride or not  so we remove Married_nan and Married_No
married_df = married_df.drop(['Married_nan','Married_No'],axis=1)
married_df

married_df.rename(columns={'Married_Yes':'Married'},inplace=True)
married_df['Married']=married_df['Married'].astype(int)
married_df  #1----> Married or 0-----> not married
```

```python
#replace the data of df['Gender'] with new encoded data 
df['Married'] = married_df['Married']
df['Married']
```

```python
df.head()
```

```python
df.isnull().sum()
#we have null values in Dependents column 
df['Dependents'] = df['Dependents'].replace({'3+': '3'})
df['Dependents'] = pd.to_numeric(df['Dependents'], errors='coerce')
df['Dependents'] = df['Dependents'].fillna(df['Dependents'].mode())
df['Dependents'] = df['Dependents'].astype(int)
df.isnull().sum()
```
